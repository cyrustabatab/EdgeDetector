```python
images = list(paths.list_images("gdrive/My Drive/edge_detector/"))
pictures = []

for image in images:
  pictures.append(cv2.imread(image,0))
```

```python
figure = plt.figure(figsize=(20,12))
for i,picture in enumerate(pictures):
  subplot = figure.add_subplot(3,3,i+1)
  subplot.imshow(picture,cmap='gray')
```

```python
def avgPixel(picture):
  average = 0
  row,col = picture.shape
  for y in picture:
    for x in y:
        average += x
  return int(average/(row*col)-30),int(average/(row*col)+30)
```

```python
figure = plt.figure(figsize=(30,20))

for i,picture in enumerate(pictures):
  subplot = figure.add_subplot(3,3,i + 1)
  low,high = avgPixel(pictures[i])
  double = cv2.resize(pictures[i],fx=2,fy=2,dsize=None)

  double = cv2.Canny(double,low,high)
  subplot.imshow(double,cmap="gray")
```
